```python
import cv2
import mediapipe as mp
import os
import uuid

# Initialize MediaPipe Pose
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

# Initialize webcam
cap = cv2.VideoCapture(0)

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break
        
        # Convert BGR to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Flip horizontally for a mirror effect
        image = cv2.flip(image, 1)
        
        # Set flag
        image.flags.writeable = False
        
        # Make detection
        results = pose.process(image)
        
        # Set flag to true
        image.flags.writeable = True
        
        # Convert RGB back to BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Render detections
        if results.pose_landmarks:
            mp_drawing.draw_landmarks(
                image,
                results.pose_landmarks,
                mp_pose.POSE_CONNECTIONS,
                mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
            )
        
        # Display the image
        cv2.imshow('Pose Estimation', image)
        
        # Optional: Save the image
        # Uncomment the following lines if you want to save frames
        # if cv2.waitKey(10) & 0xFF == ord('s'):
        #     cv2.imwrite(os.path.join('Output Images', f'pose_{uuid.uuid1()}.jpg'), image)
        
        # Break the loop
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

### Trying Both  (Merging)

```python
import cv2
import mediapipe as mp
import os
import uuid

# Initialize MediaPipe Pose and Hands
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
mp_hands = mp.solutions.hands

# Initialize webcam
cap = cv2.VideoCapture(0)

# Initialize both pose and hands models
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose, \
     mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break
        
        # Convert BGR to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Flip horizontally for a mirror effect
        image = cv2.flip(image, 1)
        
        # Set flag
        image.flags.writeable = False
        
        # Make detections
        pose_results = pose.process(image)
        hand_results = hands.process(image)
        
        # Set flag to true
        image.flags.writeable = True
        
        # Convert RGB back to BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Render pose detections
        if pose_results.pose_landmarks:
            mp_drawing.draw_landmarks(
                image,
                pose_results.pose_landmarks,
                mp_pose.POSE_CONNECTIONS,
                mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
            )
        
        # Render hand detections
        if hand_results.multi_hand_landmarks:
            for hand_landmarks in hand_results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(
                    image,
                    hand_landmarks,
                    mp_hands.HAND_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                    mp_drawing.DrawingSpec(color=(250, 44, 250), thickness=2, circle_radius=2)
                )
        
        # Display the image
        cv2.imshow('Combined Pose and Hand Estimation', image)
        
        # Optional: Save the image
        # Uncomment the following lines if you want to save frames
        # if cv2.waitKey(10) & 0xFF == ord('s'):
        #     cv2.imwrite(os.path.join('Output Images', f'combined_{uuid.uuid1()}.jpg'), image)
        
        # Break the loop
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

### Trying both (Merging with hand of the first model)

```python
import cv2
import mediapipe as mp
import numpy as np

# Initialize MediaPipe Pose and Hands
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
mp_hands = mp.solutions.hands

# Initialize webcam
cap = cv2.VideoCapture(0)

# Custom drawing function to exclude hand landmarks from pose
def draw_pose_landmarks_without_hands(image, landmarks, connections):
    # List of landmarks to exclude (hand landmarks)
    exclude_landmarks = [mp_pose.PoseLandmark.LEFT_PINKY.value, mp_pose.PoseLandmark.RIGHT_PINKY.value,
                         mp_pose.PoseLandmark.LEFT_INDEX.value, mp_pose.PoseLandmark.RIGHT_INDEX.value,
                         mp_pose.PoseLandmark.LEFT_THUMB.value, mp_pose.PoseLandmark.RIGHT_THUMB.value]

    # Draw landmarks
    for idx, landmark in enumerate(landmarks.landmark):
        if idx not in exclude_landmarks:
            cv2.circle(image, 
                       (int(landmark.x * image.shape[1]), int(landmark.y * image.shape[0])),
                       2, (245,117,66), -1)

    # Draw connections
    for connection in connections:
        start_idx, end_idx = connection
        if start_idx not in exclude_landmarks and end_idx not in exclude_landmarks:
            start_point = landmarks.landmark[start_idx]
            end_point = landmarks.landmark[end_idx]
            
            cv2.line(image, 
                     (int(start_point.x * image.shape[1]), int(start_point.y * image.shape[0])),
                     (int(end_point.x * image.shape[1]), int(end_point.y * image.shape[0])),
                     (245,66,230), 2)

# Initialize both pose and hands models
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose, \
     mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break
        
        # Convert BGR to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Flip horizontally for a mirror effect
        image = cv2.flip(image, 1)
        
        # Set flag
        image.flags.writeable = False
        
        # Make detections
        pose_results = pose.process(image)
        hand_results = hands.process(image)
        
        # Set flag to true
        image.flags.writeable = True
        
        # Convert RGB back to BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Render pose detections
        if pose_results.pose_landmarks:
            draw_pose_landmarks_without_hands(
                image,
                pose_results.pose_landmarks,
                mp_pose.POSE_CONNECTIONS
            )
        
        # Render hand detections
        if hand_results.multi_hand_landmarks:
            for hand_landmarks in hand_results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(
                    image,
                    hand_landmarks,
                    mp_hands.HAND_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                    mp_drawing.DrawingSpec(color=(250, 44, 250), thickness=2, circle_radius=2)
                )
        
        # Display the image
        cv2.imshow('Combined Pose and Hand Estimation', image)
        
        # Break the loop
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
